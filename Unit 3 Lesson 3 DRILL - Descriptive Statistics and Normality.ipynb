```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beta_dist = np.random.beta(1,25, 100)
```

```python
plt.hist(beta_dist)
plt.axvline(beta_dist.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(beta_dist.mean() + beta_dist.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
bino_dist = np.random.binomial(25,0.1, 100)
```

```python
plt.hist(bino_dist)
plt.axvline(bino_dist.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(bino_dist.mean() + bino_dist.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
chi_dist = np.random.chisquare(25, 100)
```

```python
plt.hist(chi_dist)
plt.axvline(chi_dist.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chi_dist.mean() + chi_dist.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
dir_dist = np.random.dirichlet((10,20), size = 10)
```

```python
plt.hist(dir_dist)
plt.axvline(dir_dist.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(dir_dist.mean() + dir_dist.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
geo_dist = np.random.geometric(.5, size=100)
```

```python
plt.hist(geo_dist)
plt.axvline(geo_dist.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(geo_dist.mean() + geo_dist.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
gum_dist =np.random.gumbel(loc=0.0, scale=1.0, size=100)
```

```python
plt.hist(gum_dist)
plt.axvline(gum_dist.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gum_dist.mean() + gum_dist.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Additionally:

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.

Add them together to create a third variable.
Graph the third variable using a histogram.

Compute the mean and standard deviation and plot them as vertical lines on the histogram.
Evaluate the descriptive statistics against the data.


```python
var_one = np.random.normal(5, 0.5,1000)
```

```python
var_two = np.random.normal(10,1,1000)
```

```python
var_three = var_one + var_two
```

```python
plt.axvline(var_three.mean(), color='r', linestyle='solid', linewidth=3)
plt.axvline(var_three.mean() + var_three.std(), color='r', linestyle='dashed', linewidth=3)
plt.axvline(var_three.mean() - var_three.std(), color='r', linestyle='dashed', linewidth=3)
plt.hist(var_three)
plt.show()
print("(Mean value falls in the center of cluster and probability is symmmetrical, not skewed.)")
```
